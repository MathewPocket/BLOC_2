```python
#Participant
col_to_keep = ['iid','id','gender','idg','age','field','field_cd','undergra','mn_sat','tuition','race','imprace','imprelig','from','zipcode','income','goal','date','go_out','career','career_c','sports','tvsports','exercise','dining','museums','art','hiking','gaming','clubbing','reading','tv','theater','movies','concerts','music','shopping','yoga','exphappy','expnum','match_es','satis_2','length','numdat_2','you_call','them_cal','date_3','numdat_3','num_in_3']
df_part = dataset.loc[:,col_to_keep]
df_part.drop_duplicates(inplace=True)
dataset.drop(columns=col_to_keep[1:],axis=1,inplace=True)
df_part.shape
```

```python
#questionnaires
quest = ['attr','sinc','intel','fun','amb','shar']
df_time_quest = []
for i in range(0,3):
    columns = ['iid']
    for j in range(1,8):
        for k in quest:
            #Pas d'attribut shar pour les 3 et 5
            if not(j in [3,5] and k == 'shar') and j != 6 and not(j==7 and i+1 == 1):
                item = k+str(j)+'_'+str(i+1)
                columns.append(item)
    #Stock dans un tableau les datasets pour chaque time avec les meme noms de colonnes 
    df_time_quest.append(dataset.loc[:,columns])
    df_time_quest[i].columns = df_time_quest[i].columns.str.replace('_'+str(i+1), '')
    df_time_quest[i].drop_duplicates(inplace=True)
    dataset.drop(columns=columns[1:],axis=1,inplace=True)

columns = ['iid']
for j in [1,3]:
    for k in quest:
        #Pas d'attribut shar pour les 3 et 5
        if not(j ==3 and k == 'shar') and j != 6 and not(j==7 and i+1 == 1):
            item = k+str(j)+'_s'
            columns.append(item)
df_time_quest.append(dataset.loc[:,columns])
df_time_quest[3].columns = df_time_quest[3].columns.str.replace('_'+str(i+1), '')
df_time_quest[3].drop_duplicates(inplace=True)
dataset.drop(columns=columns[1:],axis=1,inplace=True)
```

```python
#DF pour un iid avec le resultat de la rencontre
col_to_keep = ['iid','pid','position','positin1','order','partner','dec','attr','sinc','intel','fun','amb','shar','like','prob','met']
df_scorecard = dataset.loc[:,col_to_keep]
dataset.drop(columns=col_to_keep[2:],axis=1,inplace=True)
#df_scorecard['pid'] = df_scorecard['pid'].astype({'pid': 'int64'})
```

```python
#Jointure entre la decision et les infos pour récupérer le genre
df_temp = df_scorecard.loc[:,['iid','pid','dec','attr','sinc','intel','fun','amb','shar']].merge(df_part.loc[:,['iid','gender']], on='iid',how='inner')
#Rajout des infos de reponse du questionnaire d'inscription sur ce qui est important
df_temp = df_temp.merge(df_time_quest[0].loc[:,['iid','attr1','sinc1','intel1','fun1','amb1','shar1']], on='iid',how='inner')
#Récupère que les hommes avec decision a oui 
df_temp = df_temp[(df_temp['dec'] == 1) & (df_temp['gender'] == 1)]
#Mise à lechelle des score
df_temp['count_score'] = df_temp.loc[:,['attr','sinc','intel','fun','amb','shar']].sum(axis=1)
for i in ['attr','sinc','intel','fun','amb','shar']:
    df_temp[i] = df_temp[i] * 100 / df_temp['count_score']
    df_temp[i] = df_temp[i].fillna(0).astype('int64')
```
